```julia
function machineEpsilon()
    eps = 1.0
    while eps + 1.0 != 1.0
        eps /= 2.0
    end
    return eps
end

eps = machineEpsilon()
println("Machine epsilon: $eps")
```

### Newton's method: one variable



Take the equation: $F(x) = x^2 - 2 = 0$. Apply the Newton's method with $x_0 = 2, \epsilon = 10^{-15}$. Present the results such as the following:



Also plot the graph with the iterations $\times$ the value of F(x).

```julia
using Printf
using Plots

function newtonOneVariable(obj, x0, eps; maxiter=100)
    xk = x0
    f, g = obj(xk)
    k = 0
    iters = Vector{Vector{Float64}}()
    push!(iters, [k, xk, f, g])

    while abs(f) > eps && k < maxiter
        xk = xk - f / g
        f, g = obj(xk)
        k += 1
        push!(iters, [k, xk, f, g])
    end
    return xk, iters
end

function ex0703(x)
    f = x * x - 2.0
    g = 2.0 * x
    return f, g
end

root, iters = newtonOneVariable(ex0703, 2.0, 1.0e-15)

@printf("x= %g F(x)=%g\n", root, ex0703(root)[1])

```

```julia
println("k\txk\t\tF(xk)\t\tF'(xk)")
for row in iters
    @printf("%d\t%+.6E\t%+.6E\t%+.6E\n", Int(row[1]), row[2], row[3], row[4])
end

```

```julia
table = reduce(vcat, [permutedims(row) for row in iters])

plot(table[2:end, 1], table[2:end, 3],
     xlabel="Iteration",
     ylabel="F(x)",
     label="F(xk)",
     marker=:o)
```

### Algorithm 7.3: Newton's method: $n$ variables



Example: $F(x)=\left(\begin{array}{c}(x_1+1)^2+ x_2^2 - 2 \\ e^{x_1} + x_2^3 - 2 \end{array}\right)$. Run the example with $x_0= \left(\begin{array}{c} 1 \\ 1 \end{array}\right)$ and $\varepsilon = 10^{-15}$

```julia
using LinearAlgebra
using Printf
using Plots

function newtonNVariables(obj, x0, eps; maxiter=100)
    xk = x0
    f, J = obj(xk)
    k = 0
    iters = Vector{Any}()
    push!(iters, (k, copy(xk), copy(f), copy(J)))

    while norm(f) > eps && k < maxiter
        d = J \ (-f)
        xk = xk + d
        f, J = obj(xk)
        k += 1
        push!(iters, (k, copy(xk), copy(f), copy(J)))
    end

    return xk, iters
end

function ex0711(x)
    f1 = (x[1] + 1)^2 + x[2]^2 - 2
    f2 = exp(x[1]) + x[2]^3 - 2

    f = [f1, f2]

    J11 = 2 * (x[1] + 1)
    J12 = 2 * x[2]
    J21 = exp(x[1])
    J22 = 3 * x[2]^2

    J = [J11 J12;
         J21 J22]

    return f, J
end

x = [1.0, 1.0]
root, iters = newtonNVariables(ex0711, x, 1.0e-15)

f = ex0711(root)[1]
@printf("x= (%g,%g) F(x)=(%g,%g)\n", root[1], root[2], f[1], f[2])
```

```julia
println("k\txk\t\tF(xk)\t\t||F(xk)||")
for it in iters
    k, xk, fk, Jk = it
    @printf("%d\t%+.6E\t%+.6E\t%+.6E\n", k, xk[1], fk[1], norm(fk))
    @printf("  \t%+.6E\t%+.6E\n", xk[2], fk[2])
end

```

```julia
norms = [norm(it[3]) for it in iters]

plot(0:length(norms)-1, norms,
     xlabel="Iteration",
     ylabel="||F(x)||",
     label="||F(xk)||",
     marker=:o)
```

Example: $F(x)=\left(\begin{array}{c}x_1^3 - 3 x_1 x_2^2 -1 \\ x_2^3 - 3x_1^2 x_2\end{array}\right)$. Run the example with $x_0= \left(\begin{array}{c} 1 \\ 1 \end{array}\right)$ and $\varepsilon = 10^{-15}$

```julia
function ex0712(x)
    f1 = x[1]^3 - 3 * x[1] * x[2]^2 - 1.0
    f2 = x[2]^3 - 3 * x[1]^2 * x[2]

    f = [f1, f2]

    J11 = 3 * x[1]^2 - 3 * x[2]^2
    J12 = -6 * x[1] * x[2]
    J21 = -6 * x[1] * x[2]
    J22 = 3 * x[2]^2 - 3 * x[1]^2

    J = [J11 J12;
         J21 J22]

    return f, J
end
```

```julia
x = [1.0, 1.0]
root, iters = newtonNVariables(ex0712, x, 1.0e-15)

f = ex0712(root)[1]
@printf("x= (%g,%g) F(x)=(%g,%g)\n", root[1], root[2], f[1], f[2])
```

```julia
println("k\txk\t\tF(xk)\t\t||F(xk)||")
for it in iters
    k, xk, fk, Jk = it
    @printf("%d\t%+.6E\t%+.6E\t%+.6E\n", k, xk[1], fk[1], norm(fk))
    @printf("  \t%+.6E\t%+.6E\n", xk[2], fk[2])
end
```

```julia
norms = [norm(it[3]) for it in iters]

plot(0:length(norms)-1, norms,
     xlabel="Iteration",
     ylabel="||F(x)||",
     label="||F(xk)||",
     marker=:o)
```
