## Chapter 10: Newton's local method

### Algorithm 10.1: Newton's local method





Example 5.8: $f(x_1,x_2) = \frac{1}{2} x_1^2 + x_1 \cos(x_2)$; $x_0 = [1.0, 1.0]$

```julia
using LinearAlgebra
using Printf
using Plots

function newtonLocal(obj, x0, eps; maxiter=100)
    xk = copy(x0)
    f, g, H = obj(xk)
    iters = Vector{Any}()
    push!(iters, (copy(xk), f, copy(g), copy(H)))
    k = 0

    while norm(g) > eps && k < maxiter
        d = H \ (-g)
        xk = xk + d
        f, g, H = obj(xk)
        push!(iters, (copy(xk), f, copy(g), copy(H)))
        k += 1
    end

    return xk, iters
end

function ex0508(x)
    f = 0.5 * x[1]^2 + x[1] * cos(x[2])
    g = [x[1] + cos(x[2]);
         -x[1] * sin(x[2])]
    H = [1.0 -sin(x[2]);
         -sin(x[2]) -x[1] * cos(x[2])]
    return f, g, H
end

x0 = [1.0, 1.0]
sol, iters = newtonLocal(ex0508, x0, 1e-15)

println(sol)

```

```julia
println("k\txk\t\tGrad(xk)\t\t||Grad(xk)||\t\tf(xk)")
for k in eachindex(iters)
    xk, fk, gk, Hk = iters[k]
    @printf("%d\t%+.6E\t%+.6E\t%+.6E\t%+.6E\n", k - 1, xk[1], gk[1], norm(gk), fk)
    @printf("\t%+.6E\t%+.6E\n", xk[2], gk[2])
end

```

```julia
function theFunctionToPlot(x, y)
    return 0.5 .* x.^2 .+ x .* cos.(y)
end

xlist = range(-2.0, 2.0; length=1000)
ylist = range(-6.0, 6.0; length=1000)
Z = [theFunctionToPlot(x, y) for y in ylist, x in xlist]

xiter = [it[1][1] for it in iters]
yiter = [it[1][2] for it in iters]

contour(xlist, ylist, Z,
    levels=15,
    xlabel="x",
    ylabel="y",
    label="")
plot!(xiter, yiter, linewidth=3, label="iterates")

```

```julia
xlist = range(-0.5, 1.1; length=1000)
ylist = range(0.9, 2.0; length=1000)
Z = [theFunctionToPlot(x, y) for y in ylist, x in xlist]

contour(xlist, ylist, Z,
    levels=15,
    xlabel="x",
    ylabel="y",
    label="")
plot!(xiter, yiter, linewidth=3, label="iterates")
```

### Algorithm 10.2: Newton's local method by quadratic modeling



We test the algorithm on the Rosenbrock function. It is described in Section 11.6 of the book for two variables.  The implementation below involves $n$ variables: \\[f(x) = \sum_{i=1}^{n-1} f_i(x) = \sum_{i=1}^{n-1} 100 (x_{i+1}-x_i^2)^2 + (1-x_i)^2.\\]
The calculation of the derivatives is based on the partial derivatives of the functions $f_i$:
\\[
\begin{array}{rcl}
\partial f_i/\partial x_i&=& -400 x_i (x_{i+1}-x_i^2) - 2(1-x_i), \\\\
\partial f_i/\partial x_{i+1}&=& 200 (x_{i+1}-x_i^2), \\\\
\partial^2 f_i/\partial x^2_i  &=& -400 x_{i+1}+ 1200 x_i^2 + 2, \\\\
\partial^2 f_i/\partial x_i \partial x_{i+1}  &=& -400 x_i^2, \\\\
\partial^2 f_i/\partial x^2_{i+1}  &=& 200. 
\end{array}
\\]

```julia
using LinearAlgebra
using Plots

function quadraticDirect(Q, b)
    L = cholesky(Q).L
    y = L \ (-b)
    solution = L' \ y
    return solution
end

function scalarInner(a, b)
    return dot(a, b)
end

function conjugateGradient(Q, b, x0)
    n = length(x0)
    xk = copy(x0)
    gk = Q * xk + b
    iters = Vector{Any}()
    dk = -gk
    betak = 0.0
    alphak = 0.0

    for _ in 1:n
        denom = scalarInner(dk, Q * dk)
        if denom <= 0
            error("The matrix must be positive definite")
        end
        alphak = -dot(dk, gk) / denom
        push!(iters, (copy(xk), copy(gk), copy(dk), alphak, betak))
        xk = xk + alphak * dk
        gkp1 = Q * xk + b
        betak = scalarInner(gkp1, gkp1) / scalarInner(gk, gk)
        dk = -gkp1 + betak * dk
        gk = gkp1
    end

    push!(iters, (copy(xk), copy(gk), copy(dk), alphak, betak))
    return xk, iters
end

function newtonLocalQuadratic(obj, x0, eps; cg=false, maxiter=100)
    n = length(x0)
    xk = copy(x0)
    iters = Vector{Any}()
    f, g, H = obj(xk)
    push!(iters, (copy(xk), f, copy(g), copy(H)))
    k = 0

    while norm(g) > eps && k < maxiter
        if cg
            d = conjugateGradient(H, g, zeros(n))[1]
        else
            d = quadraticDirect(H, g)
        end
        xk = xk + d
        f, g, H = obj(xk)
        push!(iters, (copy(xk), f, copy(g), copy(H)))
        k += 1
    end

    return xk, iters
end

function exRosenbrock(x, hessian=false)
    n = length(x)
    f = sum(100.0 * (x[i+1] - x[i]^2)^2 + (1.0 - x[i])^2 for i in 1:n-1)

    g = zeros(n)
    for i in 1:n-1
        g[i] += -400.0 * x[i] * (x[i+1] - x[i]^2) - 2.0 * (1.0 - x[i])
        g[i+1] += 200.0 * (x[i+1] - x[i]^2)
    end

    if hessian
        H = zeros(n, n)
        for i in 1:n-1
            H[i, i] += -400.0 * x[i+1] + 1200.0 * x[i]^2 + 2.0
            H[i+1, i] += -400.0 * x[i]
            H[i, i+1] += -400.0 * x[i]
            H[i+1, i+1] += 200.0
        end
        return f, g, H
    else
        return f, g
    end
end

function exRosenbrockWithHessian(x)
    return exRosenbrock(x, true)
end

x0 = [-1.5, 2.0]
sol, iters = newtonLocalQuadratic(exRosenbrockWithHessian, x0, 1e-15)

println(sol)
```

```julia

function theFunctionToPlot(x, y)
    return 100.0 .* (y .- x.^2).^2 .+ (1.0 .- x).^2
end

xlist = range(-3.0, 3.0; length=1000)
ylist = range(-3.0, 3.0; length=1000)
Z = [theFunctionToPlot(x, y) for y in ylist, x in xlist]

xiter = [it[1][1] for it in iters]
yiter = [it[1][2] for it in iters]

contour(xlist, ylist, Z, levels=20, label="")
plot!(xiter, yiter, linewidth=3, label="iterates")
```

```julia
x0 = [-1.5, 2.0]
sol, iters = newtonLocalQuadratic(exRosenbrockWithHessian, x0, 1e-15, cg=true)

println(sol)
```

```julia

function theFunctionToPlot(x, y)
    return 100.0 .* (y .- x.^2).^2 .+ (1.0 .- x).^2
end

xlist = range(-3.0, 3.0; length=1000)
ylist = range(-3.0, 3.0; length=1000)
Z = [theFunctionToPlot(x, y) for y in ylist, x in xlist]

xiter = [it[1][1] for it in iters]
yiter = [it[1][2] for it in iters]

contour(xlist, ylist, Z, levels=20, label="")
plot!(xiter, yiter, linewidth=3, label="iterates")
```

We now apply the algorithm on example 5.8. In this case, the algorithm fails to converge, and one hessian is not positive definite. We try first using the direct method to solve the quadratic problem. An error is triggered. 

```julia
x0 = [1.1, 1.1]
sol, iters = newtonLocalQuadratic(ex0508, x0, 1e-15)
println(sol)
```

If we try with the conjugate gradient method, an error is also triggered. 

```julia
x0 = [1.1, 1.1]
sol, iters = newtonLocalQuadratic(ex0508, x0, 1e-15, cg=true)
println(sol)
```
