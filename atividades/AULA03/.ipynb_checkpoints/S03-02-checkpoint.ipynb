## Chapter 8: Quasi-Newton methods

### Algorithm 8.1: Finite difference Newton's method: one variable



Example: $F(x)=x^2-2$. Run the example with $x_0=2$ and $\tau=10^{-7}$

```julia
using Printf
using Plots

function newtonFinDiffOneVariable(obj, x0, eps, tau; maxiter=100)
    xk = x0
    f = obj(xk)[1]
    k = 0
    iters = Vector{Vector{Float64}}()
    push!(iters, [k, xk, f])

    while abs(f) > eps && k < maxiter
        s = abs(xk) >= 1 ? tau * xk : tau
        fs = obj(xk + s)[1]
        xk = xk - s * f / (fs - f)
        f = obj(xk)[1]
        k += 1
        push!(iters, [k, xk, f])
    end

    return xk, iters
end

function ex0703(x)
    f = x^2 - 2.0
    g = 2.0 * x
    return f, g
end

tau = 1.0e-7
root, iters = newtonFinDiffOneVariable(ex0703, 2.0, 1.0e-15, tau)

@printf("x= %g F(x)=%g\n", root, ex0703(root)[1])

```

```julia
println("k\txk\t\tF(xk)")
for row in iters
    @printf("%d\t%+.6E\t%+.6E\n", Int(row[1]), row[2], row[3])
end

```

```julia
table = reduce(vcat, [permutedims(row) for row in iters])

plot(table[2:end, 1], table[2:end, 3],
     xlabel="Iteration",
     ylabel="F(x)",
     label="F(xk)",
     marker=:o)
```

Run the example with $x_0=2$ and $\tau=0.1$

```julia
tau = 0.1
root, iters = newtonFinDiffOneVariable(ex0703, 2.0, 1.0e-15, tau)

@printf("x= %g F(x)=%g\n", root, ex0703(root)[1])
```



Example: $F(x)=x^2-2$, with $x_0=2$ and $a_0=1$

```julia
using Printf
using Plots

function ex0703(x)
    f = x^2 - 2.0
    g = 2.0 * x
    return f, g
end

function secantOneVariable(obj, x0, a0, eps; maxiter=100)
    xk = x0
    ak = a0
    f = obj(xk)[1]
    k = 0
    iters = Vector{Vector{Float64}}()
    push!(iters, [k, xk, f, ak])

    while abs(f) > eps && k < maxiter
        xold = xk
        xk = xk - f / ak
        fold = f
        f = obj(xk)[1]
        ak = (fold - f) / (xold - xk)
        k += 1
        push!(iters, [k, xk, f, ak])
    end

    return xk, iters
end

x0 = 2.0
a0 = 1.0
root, iters = secantOneVariable(ex0703, x0, a0, 1.0e-15)

@printf("x= %g F(x)=%g\n", root, ex0703(root)[1])

```

```julia
println("k\txk\t\tF(xk)\t\tak")
for row in iters
    @printf("%d\t%+.6E\t%+.6E\t%+.6E\n", Int(row[1]), row[2], row[3], row[4])
end

```

```julia
table = reduce(vcat, [permutedims(row) for row in iters])

plot(table[2:end, 1], table[2:end, 3],
     xlabel="Iteration",
     ylabel="F(x)",
     label="F(xk)",
     marker=:o)
```

### Algorithm 8.3: finite difference Newton's method: $n$ variables



Example: $F(x)=\left(\begin{array}{c}(x_1+1)^2+ x_2^2 - 2 \\ e^{x_1} + x_2^3 - 2 \end{array}\right)$. Run the example with $x_0= \left(\begin{array}{c} 1 \\ 1 \end{array}\right)$ and $\tau=10^{-7}$

```julia
using LinearAlgebra
using Printf
using Plots

function newtonFinDiffNVariables(obj, x0, tau, eps; maxiter=100)
    n = length(x0)
    xk = copy(x0)
    f = obj(xk)[1]
    k = 0
    iters = Vector{Any}()
    push!(iters, (k, copy(xk), copy(f)))

    I_n = Matrix{Float64}(LinearAlgebra.I, n, n)

    while norm(f) > eps && k < maxiter
        A = Matrix{Float64}(undef, n, n)

        for col in 1:n
            xi = xk[col]
            if abs(xi) >= 1
                s = tau * xi
            elseif xi >= 0
                s = tau
            else
                s = -tau
            end

            ei = I_n[:, col]
            fp = obj(xk + s * ei)[1]
            A[:, col] = (fp - f) / s
        end

        d = A \ (-f)
        xk = xk + d
        f = obj(xk)[1]
        k += 1
        push!(iters, (k, copy(xk), copy(f)))
    end

    return xk, iters
end

function ex0711(x)
    f1 = (x[1] + 1)^2 + x[2]^2 - 2
    f2 = exp(x[1]) + x[2]^3 - 2

    f = [f1, f2]

    J11 = 2 * (x[1] + 1)
    J12 = 2 * x[2]
    J21 = exp(x[1])
    J22 = 3 * x[2]^2

    J = [J11 J12;
         J21 J22]

    return f, J
end

x0 = [1.0, 1.0]
tau = 1.0e-7
root, iters = newtonFinDiffNVariables(ex0711, x0, tau, 1.0e-15)

f = ex0711(root)[1]
@printf("x= (%g,%g) F(x)=(%g,%g)\n", root[1], root[2], f[1], f[2])

```

```julia
println("k\txk\t\tF(xk)\t\t||F(xk)||")
for it in iters
    k, xk, fk = it
    @printf("%d\t%+.6E\t%+.6E\t%+.6E\n", k, xk[1], fk[1], norm(fk))
    @printf("  \t%+.6E\t%+.6E\n", xk[2], fk[2])
end

```

```julia
norms = [norm(it[3]) for it in iters]

plot(0:length(norms)-1, norms,
     xlabel="Iteration",
     ylabel="||F(x)||",
     label="||F(xk)||",
     marker=:o)
```

Run the example with $x_0= \left(\begin{array}{c} 1 \\ 1 \end{array}\right)$ and $\tau=0.1$

```julia

x0 = [1.0, 1.0]
tau = 0.1
root, iters = newtonFinDiffNVariables(ex0711, x0, tau, 1.0e-15)

f = ex0711(root)[1]
@printf("x= (%g,%g) F(x)=(%g,%g)\n", root[1], root[2], f[1], f[2])
```

```julia
norms = [norm(it[3]) for it in iters]

plot(0:length(norms)-1, norms,
     xlabel="Iteration",
     ylabel="||F(x)||",
     label="||F(xk)||",
     marker=:o)
```

### Algorithm 8.4: secant method: $n$ variables



Example 7.11: $F(x)=\left(\begin{array}{c}(x_1+1)^2+ x_2^2 - 2 \\ e^{x_1} + x_2^3 - 2 \end{array}\right)$, $x_0=\left(\begin{array}{c} 1 \\ 1 \end{array}\right)$


```julia
using LinearAlgebra
using Printf
using Plots

function ex0711(x)
    f1 = (x[1] + 1)^2 + x[2]^2 - 2
    f2 = exp(x[1]) + x[2]^3 - 2

    f = [f1, f2]

    J11 = 2 * (x[1] + 1)
    J12 = 2 * x[2]
    J21 = exp(x[1])
    J22 = 3 * x[2]^2

    J = [J11 J12;
         J21 J22]

    return f, J
end

function secantNVariables(obj, x0, eps; maxiter=100)
    n = length(x0)
    xk = copy(x0)
    f = obj(xk)[1]
    k = 0
    A = Matrix{Float64}(LinearAlgebra.I, n, n)

    iters = Vector{Any}()
    push!(iters, (k, copy(xk), copy(f), copy(A)))

    while norm(f) > eps && k < maxiter
        xold = copy(xk)
        d = A \ (-f)
        xk = xk + d
        fold = copy(f)
        f = obj(xk)[1]
        y = f - fold
        update = ((y - A * d) * transpose(d)) / dot(d, d)
        A = A + update
        k += 1
        push!(iters, (k, copy(xk), copy(f), copy(A)))
    end

    return xk, iters
end

x0 = [1.0, 1.0]
root, iters = secantNVariables(ex0711, x0, 1.0e-15)

f = ex0711(root)[1]
@printf("x= (%g,%g) F(x)=(%g,%g)\n", root[1], root[2], f[1], f[2])

```

```julia
println("k\txk\t\tF(xk)\t\t||F(xk)||")
for it in iters
    k, xk, fk, A = it
    @printf("%d\t%+.6E\t%+.6E\t%+.6E\n", k, xk[1], fk[1], norm(fk))
    @printf("  \t%+.6E\t%+.6E\n", xk[2], fk[2])
end

```

```julia
norms = [norm(it[3]) for it in iters]

plot(0:length(norms)-1, norms,
     xlabel="Iteration",
     ylabel="||F(x)||",
     label="||F(xk)||",
     marker=:o)

```

```julia
for it in iters
    k, x, f, A = it
    J = ex0711(x)[2]
    @printf("%d\t%+.6E\t%+.6E\t%+.6E\t%+.6E\n", k, J[1,1], J[1,2], A[1,1], A[1,2])
    @printf("  \t%+.6E\t%+.6E\t%+.6E\t%+.6E\n", J[2,1], J[2,2], A[2,1], A[2,2])
end
```
