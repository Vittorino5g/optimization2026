## Chapter 13: Quasi-Newton methods

### Algorithm 13.1: Quasi-Newton BFGS method





Example 5.8: $f(x_1,x_2) = \frac{1}{2} x_1^2 + x_1 \cos(x_2)$

```julia
using LinearAlgebra
using Printf
using Plots

struct OptimizationValueOutOfRange <: Exception
    msg::String
end

struct OptimizationInputError <: Exception
    msg::String
end

function Base.showerror(io::IO, e::OptimizationValueOutOfRange)
    print(io, e.msg)
end

function Base.showerror(io::IO, e::OptimizationInputError)
    print(io, e.msg)
end

function scalarInner(a, b)
    return dot(a, b)
end

function lineSearch(obj, x, d, alpha0, beta1, beta2, λ)
    if λ <= 1
        throw(OptimizationValueOutOfRange("lambda is $λ and must be > 1"))
    end
    if alpha0 <= 0
        throw(OptimizationValueOutOfRange("alpha0 is $alpha0 and must be > 0"))
    end
    if beta1 >= beta2
        throw(OptimizationValueOutOfRange("Incompatible Wolfe cond. parsmeters: beta1=$beta1 is greater than beta2=$beta2"))
    end

    f, g = obj(x)
    deriv = scalarInner(g, d)
    if deriv >= 0
        throw(OptimizationInputError("d is not a descent direction: $deriv >= 0"))
    end

    alpha = alpha0
    alphal = 0.0
    alphar = typemax(Float64)
    finished = false

    iters = Vector{Tuple{Float64,Float64,Float64,String}}()
    push!(iters, (alpha, alphal, alphar, ""))

    while !finished
        xnew = x + alpha * d
        fnew, gnew = obj(xnew)

        if fnew > f + alpha * beta1 * deriv
            reason = "too long"
            alphar = alpha
            alpha = (alphal + alphar) / 2.0
        elseif scalarInner(gnew, d) < beta2 * deriv
            reason = "too short"
            alphal = alpha
            if alphar == typemax(Float64)
                alpha = λ * alpha
            else
                alpha = (alphal + alphar) / 2.0
            end
        else
            reason = "ok"
            finished = true
        end

        push!(iters, (alpha, alphal, alphar, reason))
    end

    return alpha, iters
end

function bfgs(obj, x0; eps=1.0e-7, maxiter=100)
    n = length(x0)
    I_n = Matrix{Float64}(LinearAlgebra.I, n, n)
    Hinv = Matrix{Float64}(LinearAlgebra.I, n, n)
    xk = copy(x0)
    f, g = obj(xk)

    iters = Vector{Any}()
    push!(iters, (copy(xk), f, norm(g), copy(Hinv), 0.0))

    k = 0
    while norm(g) > eps && k < maxiter
        d = -Hinv * g
        alpha, lsIters = lineSearch(obj, xk, d, 1.0, 0.3, 0.7, 2.0)

        xold = copy(xk)
        gold = copy(g)

        xk = xk + alpha * d
        f, g = obj(xk)
        k += 1

        dbar = xk - xold
        y = g - gold
        denom = scalarInner(dbar, y)

        t1 = I_n - (dbar * y') / denom
        t2 = I_n - (y * dbar') / denom
        t3 = t1 * Hinv
        t4 = t3 * t2
        Hinv = t4 + (dbar * dbar') / denom

        push!(iters, (copy(xk), f, norm(g), copy(Hinv), denom))
    end

    return xk, iters
end

function ex0508noHessian(x)
    f = 0.5 * x[1]^2 + x[1] * cos(x[2])
    g = [x[1] + cos(x[2]);
         -x[1] * sin(x[2])]
    return f, g
end

x0 = [1.0, 1.0]
sol, iters = bfgs(ex0508noHessian, x0)

println(sol)

```

```julia
println("k\txk\t\t\t\tf(xk)\t\t||Grad(xk)||")
for (k, it) in enumerate(iters)
    xk, fk, ngk, Hinv, denom = it
    @printf("%d\t%+.6E\t%+.6E\t%+.6E\t%+.6E\n", k - 1, xk[1], xk[2], fk, ngk)
end
```

```julia

println("k\tHinv(1,1)\tHinv(2,2)\tHinv(1,2)\td'y")
for k in 2:length(iters)
    xk, fk, ngk, Hinv, denom = iters[k]
    @printf("%d\t%+.6E\t%+.6E\t%+.6E\t%+.6E\n", k - 1, Hinv[1,1], Hinv[2,2], Hinv[1,2], denom)
end
```

#### the Rosenbrock problem

```julia
function exRosenbrock(x)
    n = length(x)
    f = sum(100.0 * (x[i+1] - x[i]^2)^2 + (1.0 - x[i])^2 for i in 1:n-1)

    g = zeros(n)
    for i in 1:n-1
        g[i] += -400.0 * x[i] * (x[i+1] - x[i]^2) - 2.0 * (1.0 - x[i])
        g[i+1] += 200.0 * (x[i+1] - x[i]^2)
    end

    return f, g
end

x0 = [-1.5, 1.5]
sol, iters = bfgs(exRosenbrock, x0)

println(sol)
println("Number of iterations: $(length(iters))")
```

```julia
using Plots

function theFunctionToPlot(x, y)
    return 100.0 .* (y .- x.^2).^2 .+ (1.0 .- x).^2
end

xmin = -1.6
xmax = 1.1
xlist = range(xmin, xmax; length=1000)
ymin = -1.0
ymax = 2.0
ylist = range(ymin, ymax; length=1000)

Z = [theFunctionToPlot(x, y) for y in ylist, x in xlist]

xiter = [iters[k][1][1] for k in eachindex(iters)]
yiter = [iters[k][1][2] for k in eachindex(iters)]

contour(xlist, ylist, Z, levels=20, size=(900,600), label="")
xlims!(xmin, xmax)
ylims!(ymin, ymax)
plot!(xiter, yiter, linewidth=1, marker=:circle, label="")
scatter!([1.0], [1.0], marker=:star, label="")

```
